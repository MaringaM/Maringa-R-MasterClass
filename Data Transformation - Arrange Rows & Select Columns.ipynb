How could you use arrange() to sort all missing values to the start? (Hint: use is.na()).

```r
arrange(flights,desc(is.na(dep_time)),desc(is.na(tailnum)))
```

```r
arrange(flights,desc(dep_delay))
```

```r
arrange(flights, desc(distance))
```

What happens if you include the name of a variable multiple times in a select() call? _it will only appear once on the output_

```r
head(select(flights,arr_delay,arr_delay),5)
```

What does the one_of() function do? _It matches the variable names in the vector to the variables in the data frame when used with a select. Helps to avoid typing the variable names all the time._  

```r
vars <- c(
"year", "month", "day", "dep_delay", "arr_delay"
)
head(select(flights, one_of(vars)),5)
```

Does the result of running the following code surprise you? How do the select helpers deal with case by default? How can you change that default?  _Default is case insensitive, to change this specify ignore.case = FALSE_

```r
head(select(flights, contains("TIME")),5)
```

```r
select(flights,contains("Time", ignore.case = FALSE))
```
