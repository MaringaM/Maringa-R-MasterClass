1. Find all flights that:  
a. Had an arrival delay of two or more hours

```r
filter(flights,arr_delay>=120)
```

b.flew to Houston IAH or HOU

```r
filter(flights,dest=="IAH" | dest=="HOU")
```

```r
filter(flights, dest %in% c("IAH", "HOU"))
```

c.flew through American, United or Delta

```r
filter(flights, carrier %in% c("UA", "AA","DL"))
```

d.Departed in Summer (July, August, and September)

```r
filter(flights, month <= 9 & month >= 7)
```

e.  Arrived more than 2 hours late, but didn’t leave late

```r
filter(flights, arr_delay > 120, dep_delay <= 0)
```

f. Were delayed by at least an hour, but made up over 30 minutes in flight

```r
filter(flights, dep_delay >=60 & (dep_delay-arr_delay >= 30))
```

g. Departed between midnight and 6am (inclusive)

```r
filter(flights, dep_time ==2400 | (dep_time>0 & dep_time <= 600))
```

Another useful dplyr filtering helper is between(). What does it do? Can you use it to simplify the code needed to answer the previous challenges?  
_Checks if numeric values fall within a specified range. It's a shortcut for x>=left & x<= right_

```r
filter(flights, between(dep_time, 0, 600))
```

How many flights have a missing dep_time? What other variables are missing? What might these rows represent?

```r
count(filter(flights, is.na(dep_time)))
```

Why is NA ^ 0 not missing? _Any Value to the power of 0 is 1_  

```r
NA ^ 0 
```

Why is NA | TRUE not missing? _When TRUE is evaluated using the "OR |" operator the result is always TRUE_

```r
NA|TRUE
```

Why is FALSE & NA not missing? _When FALSE is evaluated using the "AND &" operator the result is always FALSE

```r
NA&FALSE
```
